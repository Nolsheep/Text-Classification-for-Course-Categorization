```python
import pandas as pd

# กำหนด path ของไฟล์ใน Google Drive
train_path = "/content/drive/My Drive/nlp455/workshop1/training_dataset.csv"
test_path = "/content/drive/My Drive/nlp455/workshop1/testing_dataset.csv"

# โหลดข้อมูล
train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_path)

# ลบคอลัมน์ index หากมีอยู่ในชุดทดสอบ
if 'index' in test_df.columns:
    test_df.drop(columns=['index'], inplace=True)

# แสดงตัวอย่างข้อมูล
print(train_df.head())
print(test_df.head())
```

```python
import re
from pythainlp.tokenize import word_tokenize
from pythainlp.corpus.common import thai_stopwords
from pythainlp.util import normalize

# โหลด stopwords ภาษาไทย
stopwords = set(thai_stopwords())

def preprocess_text(text):
    if not isinstance(text, str):
        return []

    # ทำ normalization (เช่น แก้คำซ้ำ ๆ เช่น "ค่ะๆ" → "ค่ะ")
    text = normalize(text)

    # แปลงเป็นตัวพิมพ์เล็ก
    text = text.lower()

    # ลบตัวอักษรที่ไม่จำเป็น (ตัวเลข, วรรคตอน, สัญลักษณ์)
    text = re.sub(r"[0-9!@#$%^&*()_+=\[\]{};:'\"\\|,.<>?/]", "", text)

    # ตัดคำโดยใช้ engine "deepcut" (แม่นยำกว่าปกติ)
    words = word_tokenize(text, engine="deepcut")

    # ลบ stopwords
    words = [word for word in words if word not in stopwords]

    return words

# ใช้กับข้อมูลเทรนและเทสต์
train_df["Name_clean"] = train_df["Name"].astype(str).apply(preprocess_text)
test_df["Name_clean"] = test_df["Name"].astype(str).apply(preprocess_text)

print(train_df[["Name", "Name_clean"]].head())
```

```python
from gensim.models import Word2Vec
import numpy as np

# โหลดโมเดล Word2Vec ที่มีการเทรนไว้แล้ว
word2vec_model = Word2Vec(sentences=train_df["Name_clean"], vector_size=300, window=5, min_count=1, workers=4)

# ฟังก์ชันแปลงคำเป็นเวกเตอร์โดยใช้ค่าเฉลี่ยแบบ Weighted
def get_sentence_vector(words, model):
    vectors = [model.wv[word] for word in words if word in model.wv]

    if not vectors:
        return np.zeros(model.vector_size)

    return np.mean(vectors, axis=0)

# แปลงข้อความเป็นเวกเตอร์
X_train = np.array([get_sentence_vector(words, word2vec_model) for words in train_df["Name_clean"]])
X_test = np.array([get_sentence_vector(words, word2vec_model) for words in test_df["Name_clean"]])

print(f"Shape of Training Data: {X_train.shape}")
print(f"Shape of Test Data: {X_test.shape}")
```

```python
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report

# แปลง Categorical เป็นตัวเลข
label_encoder = LabelEncoder()
y_train = label_encoder.fit_transform(train_df["Categorical"])

# แบ่งข้อมูลเป็น Train-Test
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# ปรับแต่งโมเดล SVM
svm_model = SVC(kernel="rbf", C=10, class_weight="balanced", probability=True)
svm_model.fit(X_train_split, y_train_split)

# ทดสอบบน Validation Set
y_val_pred = svm_model.predict(X_val)
val_acc = accuracy_score(y_val, y_val_pred)

print(f"Validation Accuracy: {val_acc:.4f}")
print(classification_report(y_val, y_val_pred, target_names=label_encoder.classes_))
```

```python
# ทำนายผลลัพธ์
y_pred = svm_model.predict(X_test)
y_pred_labels = label_encoder.inverse_transform(y_pred)

# เพิ่มคอลัมน์ y_pred ใน DataFrame
test_df["y_pred"] = y_pred_labels

# บันทึกผลลัพธ์
output_path = "/content/drive/My Drive/nlp455/workshop1/noltime.csv"
test_df[["Name", "y_pred"]].to_csv(output_path, index=False)

print(f"✅ Saved predictions to {output_path}")
```
